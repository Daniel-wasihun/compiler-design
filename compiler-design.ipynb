```python
#Program to count the number of characters,
# words, spaces, end of lines in a given input file
```

```python
%%writefile wordcount.l
%{
#include <stdio.h>
int chars = 0, words = 0, spaces = 0, lines = 0;
%}

%%

\n          { lines++; chars++; }
[ \t]       { spaces++; chars++; }
[a-zA-Z]+   { words++; chars += yyleng; }
.           { chars++; }

%%

int main() {
    yylex();
    printf("Characters: %d\nWords: %d\nSpaces: %d\nLines: %d\n", chars, words, spaces, lines);
    return 0;
}

int yywrap() {
    return 1;
}
```

```python
# Program to validate password rules
# Password > 8 characters
# Contains at least one uppercase, lowercase, digit, and special character
```

```python
%%writefile password.l
%{
#include <stdio.h>
#include <string.h>
int valid = 0;
%}

%%

[A-Za-z0-9!@#$%^&*]{8,}    {
    int has_upper = 0, has_lower = 0, has_digit = 0, has_special = 0;
    char *ptr = yytext;
    while (*ptr) {
        if (*ptr >= 'A' && *ptr <= 'Z') has_upper = 1;
        else if (*ptr >= 'a' && *ptr <= 'z') has_lower = 1;
        else if (*ptr >= '0' && *ptr <= '9') has_digit = 1;
        else if (strchr("!@#$%^&*", *ptr)) has_special = 1;
        ptr++;
    }
    if (has_upper && has_lower && has_digit && has_special) {
        valid = 1;
        printf("Valid password: %s\n", yytext);
    }
}
\n                         { if (!valid) printf("Invalid password\n"); valid = 0; }
[ \t]                      { /* ignore spaces and tabs */ }
.                          { /* ignore other characters */ }

%%

int main() {
    printf("Enter password: ");
    yylex();
    return 0;
}

int yywrap() {
    return 1;
}
```

```python
# Lex program to implement a simple calculator
```

```python
%%writefile calc.l
%{
#include <stdio.h>
float result = 0;
float operand = 0;
char last_op = '+';
%}

%%

[0-9]+(\.[0-9]+)?     {
    operand = atof(yytext);
    switch(last_op) {
        case '+': result += operand; break;
        case '-': result -= operand; break;
        case '*': result *= operand; break;
        case '/': if(operand != 0) result /= operand; else printf("Division by zero!\n"); break;
    }
}
"+"                    { last_op = '+'; }
"-"                    { last_op = '-'; }
"*"                    { last_op = '*'; }
"/"                    { last_op = '/'; }
\n                     { printf("Result: %.2f\n", result); result = 0; last_op = '+'; }
[ \t]                  { /* Ignore whitespace */ }
.                      { /* Ignore other characters */ }

%%

int main() {
    printf("Enter expression (e.g., 5 + 3): ");
    yylex();
    return 0;
}

int yywrap() {
    return 1;
}
```

```python
# Program to validate URL
```

```python
%%writefile url.l
%{
#include <stdio.h>
int valid = 0;
char buffer[1024];
int buf_pos = 0;
%}

%%

(http|https):\/\/[a-zA-Z0-9.-]+\.[a-zA-Z]{2,}(\/[a-zA-Z0-9._-]*)*\/?   {
    valid = 1;
    printf("Valid URL: %s\n", yytext);
    buf_pos = 0;
    buffer[0] = '\0';
}

[^ \t\n]+    {
    if (!valid) {
        int i = 0;
        while (yytext[i] != '\0' && buf_pos < 1023) {
            buffer[buf_pos++] = yytext[i++];
        }
        buffer[buf_pos] = '\0';
    }
}

\n           {
    if (!valid && buf_pos > 0) {
        printf("Invalid URL: %s\n", buffer);
    }
    valid = 0;
    buf_pos = 0;
    buffer[0] = '\0';
}

[ \t]        { /* Ignore whitespace */ }

%%

int main() {
    printf("Enter URL: ");
    fflush(stdout);
    yylex();
    return 0;
}

int yywrap() {
    return 1;
}
```

```python
# Program to recognize 5 well-known human language
# scripts (including Amharic/Ge’ez)
```

```python
# 1. Amharic (Ge'ez script) - Ethiopic script used in Ethiopia and Eritrea.
# 2. Latin - Used in English and many European languages.
# 3. Cyrillic - Used in Russian, Bulgarian, and other Slavic languages.
# 4. Devanagari - Used in Hindi, Sanskrit, and other Indian languages.
# 5. Arabic - Used in Arabic, Persian, and Urdu (among others).
```

```python
%%writefile script_recognizer.l
%{
#include <stdio.h>
#include <stdlib.h>

/* Counters for each script */
int amharic_count = 0;
int latin_count = 0;
int cyrillic_count = 0;
int devanagari_count = 0;
int arabic_count = 0;
int unrecognized_count = 0;
%}

/* Enable Unicode support */
%option noyywrap
%option 8bit
%option nounput
%option noinput

/* Define UTF-8 character patterns */
LATIN       [a-zA-Z]
AMHARIC     [\xE1][\x88-\x8D][\x80-\xBF]+
CYRILLIC    \xD0[\x90-\xBF]|\xD1[\x80-\x8F]+
DEVANAGARI  \xE0[\xA4-\xA5][\x80-\xBF]+
ARABIC      \xD8[\xA0-\xBF]|\xD9[\x80-\xBF]+
UTF8_OTHER  [\xC0-\xFF]+

%%

{LATIN}       { printf("Latin: %s\n", yytext); latin_count++; }
{AMHARIC}     { printf("Amharic (Ge'ez): %s\n", yytext); amharic_count++; }
{CYRILLIC}    { printf("Cyrillic: %s\n", yytext); cyrillic_count++; }
{DEVANAGARI}  { printf("Devanagari: %s\n", yytext); devanagari_count++; }
{ARABIC}      { printf("Arabic: %s\n", yytext); arabic_count++; }
{UTF8_OTHER}  { printf("Unrecognized UTF-8: %s\n", yytext); unrecognized_count++; }
.|\n          { unrecognized_count++; }

%%

int main() {
    printf("Enter text:\n");
    yylex();

    /* Print Summary */
    printf("\n Summary \n\n");
    printf("Amharic (Ge'ez) characters: %d\n", amharic_count);
    printf("Latin characters: %d\n", latin_count);
    printf("Cyrillic characters: %d\n", cyrillic_count);
    printf("Devanagari characters: %d\n", devanagari_count);
    printf("Arabic characters: %d\n", arabic_count);
    printf("Unrecognized characters: %d\n", unrecognized_count);

    return 0;
}
```

```python
!flex script_recognizer.l
!gcc lex.yy.c -o script_detector -lfl
!./script_detector < script.txt
```

```python
# Program to find the length of the longest word
```

```python
%%writefile longest_word.l
%{
#include <stdio.h>
int max_length = 0;
char longest_word[100];
%}

%%

[a-zA-Z]+   {
    if (yyleng > max_length) {
        max_length = yyleng;
        strncpy(longest_word, yytext, yyleng);
        longest_word[yyleng] = '\0';
    }
}
.|\n        { /* Ignore */ }

%%

int main() {
    printf("Longest Word Length Detector \n\n");
    fflush(stdout);
    yylex();
    if (max_length > 0) {
        printf("  Longest word: %s\n", longest_word);
        printf("  Length of longest word: %d\n", max_length);
    } else {
        printf("No valid words found.\n");
    }
    fflush(stdout);
    printf("\n End of Detection \n");
    fflush(stdout);
    return 0;
}

int yywrap() {
    return 1;
}
```

```python
# Program to identify and count positive and negative numbers
```

```python
%%writefile numbers.l
%{
#include <stdio.h>
int positive = 0, negative = 0;
%}

%%

-[0-9]+(\.[0-9]+)?     { negative++; printf("  %-10s : %s\n", "Negative", yytext); }
[0-9]+(\.[0-9]+)?      { positive++; printf("  %-10s : %s\n", "Positive", yytext); }
.|\n                   { /* Ignore */ }

%%

int main() {
    printf(" Number Classifier \n");
    yylex();
    printf("\nSummary:\n");
    printf("  Positive numbers: %d\n", positive);
    printf("  Negative numbers: %d\n", negative);
    return 0;
}

int yywrap() {
    return 1;
}
```
